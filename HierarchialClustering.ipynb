```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
data={
    'Annual Income':[15, 16, 17, 30, 31, 32, 60, 62, 65, 90, 91, 92],
    'Spending Score': [39, 40, 42, 70, 65, 68, 30, 35, 40, 85, 87, 83]
}
df = pd.DataFrame(data)

X = df[['Annual Income', 'Spending Score']].values
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
dend = dendrogram(linkage(X, method='ward'))
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
#apply hieararchial clustering
```

```python
ac=AgglomerativeClustering(n_clusters=3,linkage="ward")
y_ac=ac.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_ac, palette='Set1')
plt.title("Customer Segments")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

###############
Interpretation:
You'll see 3 clusters in different colors.

Each group represents a type of customer behavior.

Example:

Cluster 0: Low income, high spending

Cluster 1: High income, high spending

Cluster 2: Mid income, low spending
